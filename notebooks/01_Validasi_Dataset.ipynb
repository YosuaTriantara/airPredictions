```python
import pandas as pd
import numpy as np
from pathlib import Path
pd.set_option("display.max_columns", None)
```

```python
DATA_PATH = "../data/raw/station_hour.csv"
df = pd.read_csv(DATA_PATH)
```

```python
# Informasi Dataset
rows, cols = df.shape

# Validasi kolom
datetime_exists = "Datetime" in df.columns
aqi_exists = "AQI" in df.columns

# Daftar polutan utama
pollutants = [
    "PM2.5", "PM10", "NO", "NO2", "NOx",
    "NH3", "CO", "SO2", "O3",
    "Benzene", "Toluene", "Xylene"
]

available_pollutants = [
    col for col in pollutants if col in df.columns
]

# Konversi Datetime jika tersedia
datetime_type = "-"
if datetime_exists:
    df["Datetime"] = pd.to_datetime(df["Datetime"])
    datetime_type = str(df["Datetime"].dtype)

# Membuat Folder Reports
report_dir = Path("../reports")
report_dir.mkdir(parents=True, exist_ok=True)

report_path = report_dir / "dataset_validation_report.md"

# Menulis Markdown
with open(report_path, "w", encoding="utf-8") as f:

    f.write("# Dataset Validation Report\n\n")

    f.write("## Dataset\n\n")
    f.write("- File: `data/raw/station_hour.csv`\n")

    f.write("## Ukuran Dataset\n\n")
    f.write(f"- Jumlah Baris : **{rows:,}**\n")
    f.write(f"- Jumlah Kolom : **{cols}**\n\n")

    f.write("## Validasi Kolom Penting\n\n")
    f.write(f"- Datetime : {'Tersedia' if datetime_exists else 'Tidak tersedia'}\n")
    if datetime_exists:
        f.write(f"- Tipe Datetime : `{datetime_type}`\n")
    f.write(f"- AQI : {'Tersedia' if aqi_exists else 'Tidak tersedia'}\n\n")

    f.write("## Fitur Polutan Utama\n\n")
    for pol in available_pollutants:
        f.write(f"- {pol}\n")

    f.write("\n## Struktur Dataset\n\n")
    f.write("| Kolom | Tipe Data | Missing Value |\n")
    f.write("|-------|-----------|--------------:|\n")

    for col in df.columns:
        f.write(
            f"| {col} | {df[col].dtype} | {df[col].isna().sum()} |\n"
        )
        
print(f"Laporan berhasil disimpan di:\n{report_path.resolve()}")
```
