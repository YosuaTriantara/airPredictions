# KONVERSI DATETIME

```python
import pandas as pd
import os

pd.set_option("display.max_columns", None)
```

```python
DATA_PATH = "../data/raw/station_hour.csv"
df = pd.read_csv(DATA_PATH)
df[["StationId", "Datetime"]].head(20)
```

```python
df["Datetime"].dtype
```

```python
df["Datetime"] = pd.to_datetime(
    df["Datetime"],
    errors="raise"
)

print(df["Datetime"].dtype)
```

```python
print("Jumlah nilai kosong setelah parsing:")

print(df["Datetime"].isna().sum())
```

```python
df = df.sort_values(
    by=["StationId", "Datetime"]
).reset_index(drop=True)
```

```python
df[["StationId", "Datetime"]].head(20)
```

```python
is_sorted = (
    df
    .sort_values(["StationId", "Datetime"])
    .equals(df)
)

print(f"Data sudah terurut : {is_sorted}")
```

```python
OUTPUT_PATH = "../data/processed/station_hour_sorted.csv"
os.makedirs("../data/processed", exist_ok=True)

df.to_csv(
    OUTPUT_PATH,
    index=False
)

print(f"Dataset berhasil disimpan di:\n{OUTPUT_PATH}")
```
